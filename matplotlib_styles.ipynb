```python
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
```

```python
xs = np.linspace(-6, 6, 100)
yp = np.sinc(xs)
```

```python
df = pd.DataFrame({'x': xs, 'sinc': yp})
```

```python
df.plot.line(x = 'x', y = 'sinc')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-whitegrid')
df.plot.line(x = 'x', y = 'sinc')
```

```python
plt.style.use('fivethirtyeight')
df.plot.line(x = 'x', y = 'sinc')
```

```python
plt.style.use('seaborn-poster')
df.plot.line(x = 'x', y = 'sinc')
```
